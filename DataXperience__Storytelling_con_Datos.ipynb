## 1. ¿Qué Tan Bueno es tu Modelo? Valida y Evalúa tus Regresiones

Construir un modelo de Machine Learning es solo una parte del proceso; saber cuán bien funciona ese modelo es igualmente crucial. La **evaluación del desempeño** nos permite entender la precisión, robustez y generalización de nuestras predicciones. En los modelos de regresión, donde el objetivo es predecir un valor numérico continuo, existen métricas específicas y estrategias de validación que nos ayudan a cuantificar la calidad de nuestras predicciones y a evitar el sobreajuste (overfitting).



## Métricas Clave para Modelos de Regresión

Estas métricas nos ayudan a entender la diferencia entre los valores predichos por nuestro modelo y los valores reales:

*   **R² (Coeficiente de Determinación):**
    *   **Concepto:** Mide la proporción de la varianza en la variable dependiente que es predecible a partir de las variables independientes. En términos simples, nos dice qué tan bien el modelo se ajusta a los datos reales.
    *   **Interpretación:** Su valor va de 0 a 1 (o 0% a 100%). Un R² de 1 indica que el modelo explica toda la variabilidad de la variable dependiente, mientras que un 0 indica que el modelo no explica ninguna variabilidad. Generalmente, un R² más alto es mejor, pero debe interpretarse con cautela, ya que puede aumentar al añadir más variables, incluso si no son significativas.

*   **MSE (Mean Squared Error - Error Cuadrático Medio):**
    *   **Concepto:** Calcula el promedio de los cuadrados de los errores (la diferencia entre los valores predichos y los valores reales). Penaliza más fuertemente los errores grandes debido al cuadrado.
    *   **Fórmula:** $MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$
    *   **Interpretación:** Un MSE más bajo indica un mejor ajuste del modelo a los datos. La unidad del MSE es el cuadrado de la unidad de la variable dependiente, lo que puede dificultar su interpretación directa.

*   **MAE (Mean Absolute Error - Error Absoluto Medio):**
    *   **Concepto:** Calcula el promedio de los valores absolutos de los errores. Es menos sensible a los outliers que el MSE, ya que no eleva los errores al cuadrado.
    *   **Fórmula:** $MAE = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|$
    *   **Interpretación:** Un MAE más bajo indica un mejor ajuste. Su unidad es la misma que la de la variable dependiente, lo que lo hace más fácil de interpretar en el contexto del problema (ej. "el error promedio de predicción es de $5").



## Estrategias de Validación de Modelos

Para asegurar que nuestro modelo no solo funcione bien con los datos que "ya conoce", sino que también generalice a datos nuevos y no vistos, utilizamos estrategias de validación:

*   **División Train-Test:** La forma más sencilla. Se divide el conjunto de datos en dos partes: un conjunto de entrenamiento (train set) para que el modelo aprenda, y un conjunto de prueba (test set) para evaluar su desempeño en datos no vistos. La proporción común es 70/30 o 80/20.

*   **Validación Cruzada (Cross-Validation):**
    *   **Concepto:** Es una técnica más robusta que la simple división train-test. Su objetivo es evaluar la generalización del modelo a un conjunto de datos independiente, evitando el sobreajuste y ofreciendo una estimación más fiable del rendimiento del modelo.
    *   **Cómo funciona (k-fold):**
        *   El conjunto de datos se divide en $k$ "folds" (subconjuntos) de igual tamaño.
        *   El proceso se repite $k$ veces. En cada iteración:
            *   Uno de los folds se usa como conjunto de prueba.
            *   Los $k-1$ folds restantes se usan como conjunto de entrenamiento.
        *   Las métricas de desempeño se calculan para cada iteración y luego se promedian para obtener una estimación más estable del rendimiento del modelo.
    *   **¿Cuándo usarla?:** Ideal cuando se tiene un conjunto de datos de tamaño moderado y se quiere una estimación más precisa del rendimiento del modelo, reduciendo la variabilidad que podría surgir de una única división train-test.

**Nota:** La Curva ROC y el Área bajo la Curva (AUC) son métricas cruciales para la evaluación de modelos de **clasificación binaria**, no para regresión. Miden la capacidad de un modelo para distinguir entre clases, lo cual no es directamente aplicable a la predicción de valores numéricos continuos.

## 2. ¡Hola, Aprendizaje Automático! Tu Puerta de Entrada al Machine Learning

El **Machine Learning (ML)**, o Aprendizaje Automático, es una rama de la Inteligencia Artificial que permite a los sistemas aprender de los datos, identificar patrones y tomar decisiones con una mínima intervención humana. En la ciencia de datos, el ML es el motor que transforma los datos históricos en predicciones, clasificaciones y descubrimientos valiosos, abriendo un abanico de posibilidades desde la recomendación de productos hasta el diagnóstico médico.


## Aprendizaje Supervisado y No Supervisado: Las Dos Grandes Ramas

El Machine Learning se divide principalmente en dos categorías:

*   **Aprendizaje Supervisado:** Se entrena el modelo con un conjunto de datos que ya tienen "etiquetas" o respuestas correctas. El objetivo es que el modelo aprenda a predecir esas etiquetas para nuevos datos no vistos. Se utiliza para tareas de **clasificación** (predecir una categoría, ej. spam/no spam) y **regresión** (predecir un valor numérico, ej. precio de una casa).
*   **Aprendizaje No Supervisado:** Se entrena el modelo con datos que **no tienen etiquetas**. El objetivo es descubrir patrones ocultos, estructuras o relaciones dentro de los datos. Se utiliza para tareas como el **clustering** (agrupación de datos similares) o la **reducción de dimensionalidad**.


## Algoritmos Clave del Aprendizaje No Supervisado: K-Means

**K-Means:** Es un popular algoritmo de clustering. Su objetivo es agrupar un conjunto de $N$ objetos en $K$ grupos (clusters) basándose en sus características, de modo que los objetos en el mismo cluster sean más similares entre sí que los de otros clusters.

*   **Identifica patrones sin etiquetas:** Es un algoritmo de aprendizaje no supervisado. Esto significa que no necesita datos pre-etiquetados para aprender; simplemente encuentra patrones inherentes en la estructura de los datos.
*   **Ejemplo de agrupación de puntos:** Imagina que tienes una nube de puntos en un gráfico 2D. K-Means intenta encontrar $K$ centros (centroides) y asignar cada punto al centroide más cercano, formando grupos naturales. Librerías como `scikit-learn` (sklearn) en Python facilitan su implementación, y `seaborn` puede ayudarte a visualizar los clusters resultantes. Un ejemplo común es agrupar clientes por patrones de compra o segmentar imágenes.


## Algoritmos Clave del Aprendizaje Supervisado: K-Nearest Neighbors (KNN)

**KNN:** Es un algoritmo simple y efectivo utilizado para la **clasificación** y, en menor medida, para la regresión. Se basa en la idea de que los puntos de datos similares están cerca unos de otros.

*   **Clasifica nuevas observaciones con base en similitudes:** Cuando se le presenta un nuevo punto de datos, KNN busca los $K$ puntos de datos más cercanos (vecinos) en el conjunto de entrenamiento. Para clasificación, el nuevo punto se asigna a la clase más común entre sus $K$ vecinos.
*   **Ejemplo de diagrama de puntos con clasificación de color:** Imagina que tienes puntos en un gráfico, algunos azules y otros rojos. Si aparece un nuevo punto negro, KNN miraría a sus $K$ vecinos más cercanos. Si la mayoría son azules, clasificaría el punto negro como azul. Es muy intuitivo y funciona bien cuando los límites de clase son irregulares.


## Visualizando el Aprendizaje: De Datos a Patrones y Clasificaciones

La visualización es clave para entender cómo funcionan los algoritmos de ML. Piensa en un gráfico de dispersión: con K-Means, verías los puntos agrupados en diferentes colores, con un centroide por cada grupo. Con KNN, verías cómo un nuevo punto se "inclina" hacia una categoría de color dominante según sus vecinos más cercanos. Estas representaciones visuales nos ayudan a comprender la lógica detrás de las predicciones y las agrupaciones.

## Manejo de Valores Faltantes

Los valores faltantes (NaN, None, etc.) son un problema común en los datasets reales. Pueden ocurrir por diversas razones, como errores en la recolección de datos, información no disponible o fallos en el sistema. Es crucial identificar y manejar estos valores adecuadamente, ya que muchos algoritmos de Machine Learning no pueden trabajar directamente con ellos.

Existen varias estrategias para manejar los valores faltantes:

1.  **Eliminación:**
    *   **Eliminar filas:** Si una fila tiene muchos valores faltantes o si el número de filas con valores faltantes es pequeño en comparación con el tamaño total del dataset, se pueden eliminar las filas completas. Sin embargo, esto puede resultar en la pérdida de información valiosa si se eliminan demasiadas filas.
    *   **Eliminar columnas:** Si una columna tiene un porcentaje muy alto de valores faltantes (ej. > 70-80%), puede ser más sensato eliminar la columna completa, ya que no aportaría mucha información útil.

2.  **Imputación:** Reemplazar los valores faltantes con un valor estimado. Las técnicas de imputación comunes incluyen:
    *   **Media/Mediana/Moda:** Reemplazar los valores faltantes con la media (para datos numéricos sin valores atípicos), la mediana (para datos numéricos con valores atípicos) o la moda (para datos categóricos) de la columna.
    *   **Imputación por regresión:** Utilizar un modelo de regresión para predecir los valores faltantes basándose en otras columnas del dataset.
    *   **Imputación por k-NN:** Reemplazar los valores faltantes basándose en los valores de los k vecinos más cercanos.

**Ejemplo práctico: Manejo de valores faltantes en el dataset del Titanic**

Primero, identifiquemos la cantidad de valores faltantes por columna:

```python
# Contar valores faltantes por columna
print("\nValores faltantes por columna:")
print(df.isnull().sum())
```

Observamos que las columnas `Age`, `Cabin` y `Embarked` tienen valores faltantes. La columna `Cabin` tiene un porcentaje muy alto de valores faltantes, por lo que podríamos considerar eliminarla. Para `Age` y `Embarked`, podemos imputar los valores.

```python
# Estrategia 1: Eliminar la columna \'Cabin\' debido a la gran cantidad de valores faltantes
df_cleaned = df.drop('Cabin', axis=1)
print('\nDataset después de eliminar la columna Cabin:')
print(df_cleaned.head())

# Estrategia 2: Imputar valores faltantes en \'Age\' con la mediana (más robusta a outliers que la media)
df_cleaned['Age'].fillna(df_cleaned['Age'].median(), inplace=True)
print('\nValores faltantes en Age después de la imputación:')
print(df_cleaned['Age'].isnull().sum())

# Estrategia 3: Imputar valores faltantes en \'Embarked\' con la moda (el puerto de embarque más frecuente)
# Primero, encontramos la moda
most_frequent_embarked = df_cleaned['Embarked'].mode()[0]
df_cleaned['Embarked'].fillna(most_frequent_embarked, inplace=True)
print('\nValores faltantes en Embarked después de la imputación:')
print(df_cleaned['Embarked'].isnull().sum())

# Verificar que no queden valores faltantes (excepto en columnas que no se hayan tocado)
print('\nValores faltantes finales por columna:')
print(df_cleaned.isnull().sum())
```

## Manejo de Duplicados

Los registros duplicados pueden sesgar el análisis y los modelos, ya que dan un peso indebido a ciertas observaciones. Es importante identificarlos y eliminarlos si representan entradas redundantes.

```python
# Identificar filas duplicadas
duplicates = df_cleaned.duplicated()
print('\nNúmero de filas duplicadas antes de la eliminación:', duplicates.sum())

# Eliminar filas duplicadas
df_cleaned.drop_duplicates(inplace=True)
print('Número de filas duplicadas después de la eliminación:', df_cleaned.duplicated().sum())
print('Forma del dataset después de eliminar duplicados:', df_cleaned.shape)
```

## 4. Análisis y Visualización de Datos: Profundizando en las Características

Una vez que nuestros datos están limpios, el siguiente paso es comprender sus características estadísticas, especialmente la forma de sus distribuciones. Esto nos ayudará a identificar patrones, detectar anomalías y tomar decisiones informadas sobre el preprocesamiento adicional o la selección de modelos. En esta sección, aplicaremos los conceptos de distribución, asimetría, curtosis y desviación estándar a nuestro dataset del Titanic, complementado con diversas visualizaciones.


## Medidas Estadísticas Clave

Revisaremos las medidas de tendencia central y dispersión, y luego nos enfocaremos en la asimetría y curtosis.

```python
# Medidas de tendencia central y dispersión para columnas numéricas
print("\nMedidas estadísticas clave para columnas numéricas:")
print(df_cleaned[["Age", "Fare", "Pclass", "SibSp", "Parch"]].agg(["mean", "median", "std", "min", "max"]))

# Moda para columnas numéricas (puede haber múltiples modas)
print("\nModa para columnas numéricas:")
for col in ["Age", "Fare", "Pclass", "SibSp", "Parch"]:
    print(f"Moda de {col}: {df_cleaned[col].mode().tolist()}")

# Asimetría (Skewness)
print("\nAsimetría (Skewness):")
print(df_cleaned[["Age", "Fare", "Pclass", "SibSp", "Parch"]].skew())

# Curtosis (Kurtosis)
print("\nCurtosis (Kurtosis):")
print(df_cleaned[["Age", "Fare", "Pclass", "SibSp", "Parch"]].kurtosis())
```

**Interpretación:**

*   **Media, Mediana, Moda:** Comparar estos valores nos da una idea de la simetría de la distribución. Si son muy diferentes, sugiere asimetría.
*   **Desviación Estándar y Rango:** Indican la dispersión de los datos. Una desviación estándar alta significa que los datos están más dispersos alrededor de la media.
*   **Asimetría:** Valores positivos indican una cola a la derecha (más valores altos), negativos una cola a la izquierda (más valores bajos).
*   **Curtosis:** Valores positivos indican un pico más pronunciado y colas más pesadas (leptocúrtica), negativos un pico más plano y colas más ligeras (platicúrtica).

## Visualización de Distribuciones

Utilizaremos histogramas, gráficos de caja y bigotes (box plots) y gráficos de violín para visualizar las distribuciones de las columnas numéricas.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración para mejorar la visualización
sns.set_style("whitegrid")

# Columnas numéricas para visualización
numeric_cols = ["Age", "Fare", "SibSp", "Parch"]

# Histograma para cada columna numérica
plt.figure(figsize=(15, 10))
for i, col in enumerate(numeric_cols):
    plt.subplot(2, 2, i + 1)
    sns.histplot(df_cleaned[col], kde=True, bins=30)
    plt.title(f"Distribución de {col}")
    plt.xlabel(col)
    plt.ylabel("Frecuencia")
plt.tight_layout()
plt.show()

# Gráficos de Caja y Bigotes (Box Plots) para cada columna numérica
plt.figure(figsize=(15, 10))
for i, col in enumerate(numeric_cols):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(y=df_cleaned[col])
    plt.title(f"Box Plot de {col}")
    plt.ylabel(col)
plt.tight_layout()
plt.show()

# Gráficos de Violín (Violin Plots) para cada columna numérica
plt.figure(figsize=(15, 10))
for i, col in enumerate(numeric_cols):
    plt.subplot(2, 2, i + 1)
    sns.violinplot(y=df_cleaned[col])
    plt.title(f"Violin Plot de {col}")
    plt.ylabel(col)
plt.tight_layout()
plt.show()
```

**Interpretación de los gráficos:**

*   **Histogramas:** Muestran la forma general de la distribución, la presencia de picos y la asimetría. La curva KDE (Kernel Density Estimate) superpuesta ayuda a visualizar la densidad de probabilidad.
*   **Gráficos de Caja y Bigotes (Box Plots):** Son excelentes para visualizar la mediana, los cuartiles (Q1 y Q3), el rango intercuartílico (IQR) y la presencia de valores atípicos (outliers). La "caja" representa el 50% central de los datos, y los "bigotes" se extienden hasta 1.5 veces el IQR desde los cuartiles. Los puntos fuera de los bigotes son considerados atípicos.
*   **Gráficos de Violín (Violin Plots):** Combinan las características de un box plot con una estimación de la densidad de probabilidad (similar a un histograma suavizado). La "forma" del violín muestra la distribución de los datos, siendo más ancha donde hay mayor densidad de puntos. Son útiles para comparar distribuciones entre diferentes categorías o para visualizar la forma de la distribución junto con la mediana y los cuartiles.

## Manejo de Valores Faltantes en R

En R, el manejo de valores faltantes (`NA`) es similar al de Python. Identificaremos y trataremos los valores faltantes en las columnas `Age`, `Cabin` y `Embarked`.

```python
python
%%R
# Contar valores faltantes por columna
cat("\nValores faltantes por columna en R:\n")
print(colSums(is.na(df_r)))

# Estrategia 1: Eliminar la columna \'Cabin\'
df_r_cleaned <- subset(df_r, select = -c(Cabin))
cat("\nDataset después de eliminar la columna Cabin en R:\n")
print(head(df_r_cleaned))

# Estrategia 2: Imputar valores faltantes en \'Age\' con la mediana
df_r_cleaned$Age[is.na(df_r_cleaned$Age)] <- median(df_r_cleaned$Age, na.rm = TRUE)
cat("\nValores faltantes en Age después de la imputación en R:\n")
print(sum(is.na(df_r_cleaned$Age)))

# Estrategia 3: Imputar valores faltantes en \'Embarked\' con la moda
# La moda en R para factores se puede encontrar con table y which.max
mode_embarked <- names(which.max(table(df_r_cleaned$Embarked)))
df_r_cleaned$Embarked[is.na(df_r_cleaned$Embarked)] <- mode_embarked
cat("\nValores faltantes en Embarked después de la imputación en R:\n")
print(sum(is.na(df_r_cleaned$Embarked)))

# Verificar que no queden valores faltantes
cat("\nValores faltantes finales por columna en R:\n")
print(colSums(is.na(df_r_cleaned)))
```

## Análisis de Distribución, Asimetría y Curtosis en R

Para calcular la asimetría y la curtosis en R, a menudo se utiliza el paquete `e1071`. Visualizaremos las distribuciones con histogramas.

```python
# Instalar y cargar la librería e1071 si no está instalada
if (!require(e1071)) {
  install.packages("e1071")
  library(e1071)
}

# Histograma de la Edad
hist(df_r_cleaned$Age, main="Distribución de la Edad de los Pasajeros (R)",
     xlab="Edad", ylab="Frecuencia", col="skyblue", border="black")

# Histograma de la Tarifa
hist(df_r_cleaned$Fare, main="Distribución de la Tarifa Pagada (R)",
     xlab="Tarifa", ylab="Frecuencia", col="lightgreen", border="black")

# Calcular la asimetría
cat("\nAsimetría (Skewness) en R:\n")
print(skewness(df_r_cleaned$Age))
print(skewness(df_r_cleaned$Fare))

# Calcular la curtosis
cat("\nCurtosis (Kurtosis) en R:\n")
print(kurtosis(df_r_cleaned$Age))
print(kurtosis(df_r_cleaned$Fare))
```

## Normalización y Estandarización en R

En R, podemos realizar normalización y estandarización utilizando funciones base o paquetes como `caret`.

```python
python
%%R
# Normalización (Min-Max Scaling) para Fare
min_val <- min(df_r_cleaned$Fare)
max_val <- max(df_r_cleaned$Fare)
df_r_cleaned$Fare_Normalized <- (df_r_cleaned$Fare - min_val) / (max_val - min_val)

cat("\nPrimeras 5 filas de la columna Fare Normalizada en R:\n")
print(head(df_r_cleaned[, c("Fare", "Fare_Normalized")]))

# Estandarización (Z-score Scaling) para Fare
mean_val <- mean(df_r_cleaned$Fare)
sd_val <- sd(df_r_cleaned$Fare)
df_r_cleaned$Fare_Standardized <- (df_r_cleaned$Fare - mean_val) / sd_val

cat("\nPrimeras 5 filas de la columna Fare Estandarizada en R:\n")
print(head(df_r_cleaned[, c("Fare", "Fare_Standardized")]))

# Visualizar las distribuciones transformadas en R
par(mfrow=c(1, 2)) # Configurar para 1 fila, 2 columnas de gráficos
hist(df_r_cleaned$Fare_Normalized, main="Tarifa Normalizada (R)", xlab="", col="lightblue", border="black")
hist(df_r_cleaned$Fare_Standardized, main="Tarifa Estandarizada (R)", xlab="", col="lightcoral", border="black")
```
